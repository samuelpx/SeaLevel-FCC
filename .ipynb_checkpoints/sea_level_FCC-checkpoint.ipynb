```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
def draw_plot():
    # Read data from file
    df = pd.read_csv("epa-sea-level.csv")


    # Create scatter plot
    plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])

# Get slope, intercept, r-value, p-value, and standard error of the regression line
    slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

# Create line of best fit
    line = slope * df['Year'] + intercept

# Plot line of best fit
    plt.plot(df['Year'], line, color='r', label='Line of Best Fit')# Predict sea level rise in 2050
    year_2050 = 2050
    sea_level_2050 = slope * year_2050 + intercept
    plt.scatter(year_2050, sea_level_2050, color='g', label='Predicted Sea Level in 2050')

# Get data from year 2000 onwards
    df_recent = df[df['Year'] >= 2000]

# Get slope, intercept, r-value, p-value, and standard error of the recent regression line
    slope_recent, intercept_recent, r_value_recent, p_value_recent, std_err_recent = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])

# Create line of best fit for recent data
    line_recent = slope_recent * df_recent['Year'] + intercept_recent

# Plot line of best fit for recent data
    plt.plot(df_recent['Year'], line_recent, color='b', label='Line of Best Fit (2000-Current)')

# Predict sea level rise in 2050 using recent data
    sea_level_2050_recent = slope_recent * year_2050 + intercept_recent
    plt.scatter(year_2050, sea_level_2050_recent, color='m', label='Predicted Sea Level in 2050 (2000-Current)')

# Set labels and title
    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')

# Add legend
    plt.legend()


    
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```

```python
draw_plot()
```
